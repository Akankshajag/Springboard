```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas.util.testing as tm
%matplotlib inline
```

```python
#Reading the data file
hf=pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
# From the above output can be declared that columns anaemia, diabetes, high_blood_pressure, sex, smoking and death-event
#are binary columns and where as colums age, creatinine_phosphokinase, ejection_fraction, platelets, serum_creatinine and
#serum_sodium are numerical (continuous) columns.
```

```python
#Function to plot categorical variable with respect to the target value(death_rate)

col=['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking']
fig,ax = plt.subplots(2,3,figsize=(15,10))
for i in enumerate(col):
    plt.subplot(2,3,i[0]+1)
    sns.countplot(i[1], hue='death_event', data=hf, palette='RdPu' )
```

```python
#Observation from the above graph.
#1. Death event is nearly same whether the person is anaemic or not.
#2. Death event is higer in the male gender when compared to the female.
```

```python
#Function to display the distribution of the continuos varibale.

col1= ['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine','serum_sodium']

fig,ax = plt.subplots(2,3,figsize=(15,10))
for i in col1:
    index = col1.index(i)
    plt.subplot(2,3,index+1)
    sns.distplot(hf[i], color='#0DC11C')
```

```python
# The normal range for creatinine phophokinase is 10 to 120 micrograms per liter (mcg/L).
# A normal ejection_fraction reading for adults over 20 years of age is 53 to 73 percent. 
# Below 53 percent for women and 52 percent for men is considered low.
# The normal range for creatinine in the blood may be 0.84 to 1.21 milligrams per deciliter.
# A normal blood sodium level is between 135 and 145 milliequivalents per liter (mEq/L)
```

```python
# 2x2 matrix between the numerical value.

df= hf[['age', 'creatinine_phosphokinase', 'ejection_fraction', 'platelets', 'serum_creatinine','serum_sodium']]
x=df.corr()
x
```

```python
plt.figure(figsize=(15,6))
corr = x.corr()
sns.heatmap(corr,annot=True, cmap='vlag')
```

```python
# Heatmap states positivie relation between 
# 1. ejection_fraction and serum_sodium
# 2. age and serum_creatinine
# 3. serum_creatinine and serum_sodium (Negative)
```

```python
# Line graph to understand relation between ejection_fraction and serum_sodium
sns.set(rc={'figure.figsize':(20,5)})
sns.lmplot(x='serum_sodium', y='ejection_fraction', data=hf, ci=None, hue='death_event' )
```
